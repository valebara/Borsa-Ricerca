```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_Hypercolumn_8ori(mfr, n):

    titles = ['0°','30°','45°','60°','90°','120°','135°','150°']
    nL = n**2
    
    for ii in range(8):
        
        a = np.asarray(mfr[nL*ii : nL*(ii+1)])
        b = a.reshape(n,n)
        
        plt.subplot(2,4,ii+1)
        plt.imshow(b,cmap='gray',vmin=-0.001, vmax=0.02)
        plt.xticks([])
        plt.yticks([])
        plt.title(titles[ii])
        
mfr = np.load('mfr_mean.npy',allow_pickle=True)

#which digit to show
digit = 1
```

# 50% quantile

```python
mfr_mean = np.zeros(10,dtype=object)

q = np.zeros(10)

B = np.zeros((800,10))

for ind in range(10):
    
    A = np.zeros(800)
    tmp = mfr[ind]
    
    m = np.quantile(mfr[ind],.5)
    q[ind] = m
    
    A[tmp>=m] = tmp[tmp>=m]/sum(tmp[tmp>=m])
    A[tmp<m] = -(tmp[tmp<m]-m)/sum(tmp[tmp<m]-m)
    
    mfr_mean[ind] = A
    B[:,ind] = A

a = mfr_mean[digit]

print(f'Threshold: {np.mean(q):.2f}')
visualize_Hypercolumn_8ori(a, 10)
    
np.save('mfr_th_50', B)
```
