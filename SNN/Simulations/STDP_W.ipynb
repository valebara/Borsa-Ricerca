From 10x10x8 complex cells to each digit (classifications)

```python
from brian2 import *
%matplotlib inline
import numpy as np

def visualize_Hypercolumn_8ori(mfr, n=10):

    titles = ['0°','30°','45°','60°','90°','120°','135°','150°']
    nL = n**2
    
    for ii in range(8):
        
        a = asarray(mfr[nL*ii : nL*(ii+1)])
        b = a.reshape(n,n)
        
        plt.subplot(2,4,ii+1)
        plt.imshow(b,cmap='gray')
        plt.xticks([])
        plt.yticks([])
        plt.title(titles[ii])
        
w = np.zeros((800,10))
        
for ii in range(10):
    filename = 'w'+str(ii)+'.npy'
    tmp = np.load(filename)
    w[:,ii] = tmp[:,-1]

np.save('w_stdp',w)
```

```python
visualize_Hypercolumn_8ori(w[:,0])
```
