```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_Hypercolumn_8ori(mfr, n):

    titles = ['0°','30°','45°','60°','90°','120°','135°','150°']
    nL = n**2
    
    for ii in range(8):
        
        a = np.asarray(mfr[nL*ii : nL*(ii+1)])
        b = a.reshape(n,n)
        
        plt.subplot(2,4,ii+1)
        plt.imshow(b,cmap='gray')
        plt.xticks([])
        plt.yticks([])
        plt.title(titles[ii])
        
mfr = np.load('all_MFR.npy',allow_pickle=True)
t_train = np.load('t_train.npy')
```

```python
ind = 0
visualize_Hypercolumn_8ori(mfr[ind], 10)
print(f'Real digit: {t_train[ind]}')
```
