```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_Hypercolumn_8ori(mfr, n):

    titles = ['0°','30°','45°','60°','90°','120°','135°','150°']
    nL = n**2
    
    for ii in range(8):
        
        a = np.asarray(mfr[nL*ii : nL*(ii+1)])
        b = a.reshape(n,n)
        
        plt.subplot(2,4,ii+1)
        plt.imshow(b,cmap='gray')
        plt.xticks([])
        plt.yticks([])
        plt.title(titles[ii])
        
mfr = np.load('Data Complex 3x3/MFR (digit)/mfr_mean.npy',allow_pickle=True)

mfr_mean = np.zeros(10,dtype=object)

for ind in range(10):
    
    A = np.zeros(800)
    tmp = mfr[ind]
    
    m = np.mean(mfr[ind])
    
    A[tmp>=m] = tmp[tmp>=m]/sum(tmp[tmp>=m])
    A[tmp<m] = -(tmp[tmp<m]-m)/sum(tmp[tmp<m]-m)
    
    #A[tmp<m] = (tmp[tmp<m]-m)/sum(tmp[tmp<m]) v0
    
    mfr_mean[ind] = A
```

```python
B = np.zeros((800,10))

for i in range(10):
    B[:,i] = mfr_mean[i]
    
np.save('mfr_th_mean', B)
```

```python
visualize_Hypercolumn_8ori(mfr_mean[0], 10)
```
