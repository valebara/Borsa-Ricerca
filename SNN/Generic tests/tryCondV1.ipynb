```python
import matplotlib.pyplot as plt
import numpy as np

n = 28

mat = np.zeros((28,28))

[X1,Y1] = np.meshgrid(range(n), range(n))
[X2,Y2] = np.meshgrid(range(n), range(n))

x_target = 12
y_target = 20

for x in range(n):
    for y in range(n):
        
        lengthBar = abs(X1[x,y]-x_target)<=5 and abs(Y1[x,y]-y_target)<=5
        cond = lengthBar and X1[x,y]-x_target+Y1[x,y]-y_target<=2 and X1[x,y]-x_target+Y1[x,y]-y_target>=-2
        
        if cond:
            mat[x, y] = 1

plt.imshow(mat,cmap='gray')
```
