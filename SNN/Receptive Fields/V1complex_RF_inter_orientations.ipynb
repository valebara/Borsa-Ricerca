```python
import matplotlib.pyplot as plt
import numpy as np

n=28
nc=10
a =[]

[x_pre, y_pre] = np.meshgrid(range(n),range(n))
[x_post, y_post] = np.meshgrid(range(nc),range(nc))

x_post = 4
y_post = 4
domain = np.logical_and(abs(3*x_post-x_pre)<=1 , abs(3*y_post-y_pre)<=1)
angles = np.array([0, 1/6, 1/4, 1/3, 1/2, 2/3, 3/4, 5/6])*np.pi

for i in range(8):
    y = np.cos(2*(angles-angles[i]))
    a.append(sum(y>=0.49))

    if i==0:
        plt.plot(angles, y)
        plt.xticks(ticks=angles, labels=[0,30,45,60,90,120,135,150])
        plt.show()
        
        print(y)
```
